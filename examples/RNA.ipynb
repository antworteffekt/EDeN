#RNA family characterization

Application scenario: we want to characterize RNA family A with respect to RNA family B, in other words, we are interested in identifying regions with their structural contexts that are characteristic of family A but not family B

```python
%pylab inline
import pylab as pl
```

Data acquisition

```python
#given the RFAM id of a family we retrieve it from the RFAM online database
rf='RF00162'
rf='RF00005'
#we compose the correspondent URL
url_string='http://rfam.xfam.org/family/%s/alignment?acc=%s&format=fastau&download=0'%(rf,rf)

#we use an iterator that converts from FASTA to the k-representative shapes graph 
from eden.converter.RNA import RNA_SHAPE
pre_g_it = RNA_SHAPE.RNA_SHAPE_to_eden(input = url_string, input_type = 'url' , 
                                       options = {'shape_type':5,'energy_range_%':15, 'max_num_shapes':2})

from eden.modifier.graph import vertex_attributes as g_va
w_g_it = g_va.trapezoidal_reweighting(pre_g_it, 
                                             high_weight = 1.0, 
                                             low_weight = 0.01, 
                                             high_weight_window_start = 50, 
                                             high_weight_window_end = 60, 
                                             low_weight_window_start = 40, 
                                             low_weight_window_end = 70)

from eden.modifier.RNA import vertex_attributes as RNA_va
s_g_it = RNA_va.add_paired_unpaired_vertex_type(w_g_it)

g_it = g_va.colorize(s_g_it,labels=["A","U","o","G","C"])

#for sanity check we iterate on a handful (num) of seqeunces to display the resulting graph
num = 1
from eden.util import display
import itertools as it
for g in it.islice(g_it,num):
    display.draw_graph(g, size=20, node_border = False, node_size = 400, 
                       secondary_vertex_label = 'type',
                       vertex_color = 'level', colormap = 'bwr', invert_colormap = False)
```

Vectorization setup

```python
#we set the 'resolution' of the vectorization
#the radius parameter is metaphorically equivalent to the pixel resolution in an image 
max_radius=2
#the distance parameter is metaphorically equivalent to the number of shades of colors in an image
max_distance=4
#a vectorizer object can then be instanciated
from eden import graph
vec=graph.Vectorizer(r=max_radius,d=max_distance)
```

Vectorization

```python
%%time
#we reset the iterator
rf='RF00162'
url_string='http://rfam.xfam.org/family/%s/alignment?acc=%s&format=fastau&download=0'%(rf,rf)
g_it = RNA_SHAPE.RNA_SHAPE_to_eden(input = url_string, input_type = 'url')
#we employ the vectorizer to transform the graph instances into sparse vectors
#we collect all the vectors in a data matrix 
#depending on the number of instances and the desired resolution, the procedure can take from seconds to minutes for a few hundreds of sequences
X1=vec.transform(g_it, n_jobs=1)
print 'Instances: %d ; Features: %d with an avg of %d features per instance' % (X1.shape[0], X1.shape[1],  X1.getnnz()/X1.shape[0])
```

```python
%%time
#we repeat the procedure for the second RNA family
rf='RF01725'
url_string='http://rfam.xfam.org/family/%s/alignment?acc=%s&format=fastau&download=0'%(rf,rf)
g_it = RNA_SHAPE.RNA_SHAPE_to_eden(input = url_string, input_type = 'url')
X2=vec.transform(g_it, n_jobs=1)
print 'Instances: %d ; Features: %d with an avg of %d features per instance' % (X2.shape[0], X2.shape[1],  X2.getnnz()/X2.shape[0])
```

```python
#we concatenate the two data matrices into one matrix
#and we create a target vector that identifies with +1 and -1 the instances from the first or the second RNA family

from scipy.sparse import vstack
import numpy as np

#create target array
yp = [1] * X1.shape[0]
yn = [-1] * X2.shape[0]
y = np.array(yp + yn)
#update data matrix
X = vstack( [X1,X2] , format = "csr")
print 'Instances: %d ; Features: %d with an avg of %d features per instance' % (X.shape[0], X.shape[1],  X.getnnz()/X.shape[0])
```

Fit a predictor

```python
%%time
#induce a stochastic gradient descent SVM predictive model
from sklearn.linear_model import SGDClassifier
predictor = SGDClassifier()

#assess the generalization capacity of the model via a 10-fold cross validation
from sklearn import cross_validation
scores = cross_validation.cross_val_score(predictor, X, y,cv=10, scoring='roc_auc')

import numpy as np
print('AUC ROC: %.4f +- %.4f' % (np.mean(scores),np.std(scores)))
#AUC ROC values above 0.8 are indicative of a good generalization capacity 
```

```python
#fit the predictor on all data
predictor.fit(X,y)
```

Setup the Annotator

```python
#the Annotator takes in input a vecotrizer and a predictor
iann=graph.Annotator(estimator = predictor, vectorizer = vec)
```

```python
#reset the FASTA to graph converter to the first RNA family
rf='RF00162'
url_string='http://rfam.xfam.org/family/%s/alignment?acc=%s&format=fastau&download=0'%(rf,rf)
g_it = RNA_SHAPE.RNA_SHAPE_to_eden(input = url_string, input_type = 'url')
```

```python
#annotate with colors the discriminative importance of different regions of the structural graphs
start = 0
stop = 5
step = 1
for g in it.islice(iann.transform(g_it),start,stop,step):
    display.draw_graph(g, vertex_color='importance', size = 20, node_size = 200, node_border=False, colormap='RdYlGn')
```
