```python
#parameters setting
max_radius=1
max_distance=0

from eden import graph
vec=graph.Vectorizer(r=max_radius,d=max_distance)
```

```python
%%time
rf='RF00162'
url_string='http://rfam.xfam.org/family/%s/alignment?acc=%s&format=fastau&download=0'%(rf,rf)
g_it = RNA_MFE.RNA_MFE_to_eden(input = url_string, input_type = 'url')
X1=vec.transform(g_it, n_jobs=1)
print 'Instances: %d Features: %d with an avg of %d features per instance' % (X1.shape[0], X1.shape[1],  X1.getnnz()/X1.shape[0])
```

```python
%%time
rf='RF01725'
url_string='http://rfam.xfam.org/family/%s/alignment?acc=%s&format=fastau&download=0'%(rf,rf)
g_it = RNA_MFE.RNA_MFE_to_eden(input = url_string, input_type = 'url')
X2=vec.transform(g_it, n_jobs=1)
print 'Instances: %d Features: %d with an avg of %d features per instance' % (X2.shape[0], X2.shape[1],  X2.getnnz()/X2.shape[0])
```

```python
from scipy.sparse import vstack
import numpy as np

#create target array
yp = [1] * X1.shape[0]
yn = [-1] * X2.shape[0]
y = np.array(yp + yn)
#update data matrix
X = vstack( [X1,X2] , format = "csr")
print 'Instances: %d - Features: %d with an avg of %d features per instance' % (X.shape[0], X.shape[1],  X.getnnz()/X.shape[0])
```

```python
%%time
#induce a predictive model
from sklearn.linear_model import SGDClassifier
predictor = SGDClassifier()

from sklearn import cross_validation
scores = cross_validation.cross_val_score(predictor, X, y,cv=10, scoring='roc_auc')

import numpy as np
print('AUC ROC: %.4f +- %.4f' % (np.mean(scores),np.std(scores)))
```

```python
predictor.fit(X,y)
iann=graph.Annotator(estimator = predictor, vectorizer = vec)
```

```python
rf='RF00162'
url_string='http://rfam.xfam.org/family/%s/alignment?acc=%s&format=fastau&download=0'%(rf,rf)
g_it = RNA_MFE.RNA_MFE_to_eden(input = url_string, input_type = 'url')
```

```python
for g in it.islice(iann.transform(g_it),5):
    display.draw_graph(g, vertex_color='weight',size=11, node_size=200, node_border=False)
```
