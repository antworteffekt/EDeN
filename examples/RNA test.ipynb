```python
import random

def random_sequence(size = None, list_of_chars = None):
    num_chars = len(list_of_chars)
    seq = []
    #compose a list of repreated occurrences of characters in list_of_chars
    for c in list_of_chars:
        seq += [c]*(size/num_chars)
    #shuffle the list
    random.shuffle(seq)
    return ''.join(seq)
```

```python
dim = 1
seqs = []
for i in range(dim):
    seq = random_sequence(70, ['A','C','G','U'])
    seqs += ['>ID_%s'%(i)]
    seqs += [seq]
    print seq
```

```python
options = {'mode':'RNAfold', 
           'wins':[20,40], 
           'absolute_shift':[10,20]}
g_it = RNA_sequence.RNA_sequence_to_eden(input = seqs, input_type = 'list', options = options)
```

```python
from eden.util import display
vertex_label='label'
for g in g_it:
    print display.serialize_graph(g)
    display.draw_graph(g, size=12, node_size=500, vertex_label=vertex_label)
```
