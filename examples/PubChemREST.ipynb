```python
PROLOG='https://pubchem.ncbi.nlm.nih.gov/rest/pug/'
#select a assay ID
AID = str(640)
RESTQ = PROLOG + 'assay/aid/' + AID + '/sids/JSON?sids_type=active&list_return=listkey'
reply=requests.get(RESTQ)
```

```python
#extract the listkey
listkey = reply.json()['IdentifierList']['ListKey']
size = reply.json()['IdentifierList']['Size']
print 'Answer comprises %s instances'%size
```

```python
fname='dat.sdf'
DATA=[]
stepsize=100
index_start=0
for chunk, index_end in enumerate(range(0,size,stepsize)):
    if index_end is not 0 :
        print 'Chunk %s) Processing compounds %s to %s (of a total of %s)' % (chunk, index_start, index_end-1, size)
        RESTQ = PROLOG + 'substance/listkey/' + str(listkey) + '/SDF?&listkey_start=' + str(index_start) + '&listkey_count=' + str(stepsize)
        reply=requests.get(RESTQ)
        DATA+=[reply.text]
    index_start = index_end
with open(fname,'w') as file_handle:
    file_handle.write(''.join(DATA)) 
```
