```python
vec=graph.Vectorizer(r = 1, d = 0, discretization_size = 4, discretization_dimension = 7)
X=vec.fit_transform(G_list)
print 'Instances: %d Features: %d with an avg of %d features per instance' % (X.shape[0], X.shape[1],  X.getnnz()/X.shape[0])
```
