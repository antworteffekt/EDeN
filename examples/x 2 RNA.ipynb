given the RFAM id of a family we retrieve it from the RFAM online database
composing the correspondent URL

```python
def rfam_url(family_id):
    return 'http://rfam.xfam.org/family/%s/alignment?acc=%s&format=fastau&download=0'%(family_id,family_id)

def rfam_info(family_id):
    return 'http://rfam.xfam.org/family/%s#tabview=tab3'%(family_id.lower())
```

```python
def pre_process(data):
    from eden.modifier.fasta import fasta_to_fasta
    seqs = fasta_to_fasta(data)
    
    from eden.converter.fasta import fasta_to_eden
    graphs = fasta_to_eden(seqs)

    from eden.modifier.graph import vertex_attributes 
    graphs = vertex_attributes.colorize(graphs, output_attribute = 'level', labels = ['A','U','C','G'])
    
    return graphs
```

```python
def pre_process(data):
    from eden.converter.rna.rnashapes import rnashapes_to_eden
    graphs = rnashapes_to_eden(data, shape_type=5, energy_range=30, max_num=3)
    
    #annotate in node attribute 'type' the incident edges' labels
    from eden.modifier.graph import vertex_attributes
    graphs = vertex_attributes.incident_edge_label(graphs, level = 1, output_attribute = 'type', separator = '.')
    from eden.modifier.graph.structure import contraction, contraction_modifier
    #reduce all 'label' attributes of contracted nodes to a histogram to be written in the 'label' attribute of the resulting graph 
    label_modifier = contraction_modifier(attribute_in='type', attribute_out='label', reduction='set_categorical')
    #reduce all 'weight' attributes of contracted nodes using a sum to be written in the 'weight' attribute of the resulting graph 
    weight_modifier = contraction_modifier(attribute_in='weight', attribute_out='weight', reduction='sum')
    modifiers = [label_modifier, weight_modifier]
    #contract the graph on the 'type' attribute
    graphs = contraction(graphs, contraction_attribute = 'type', modifiers = modifiers, nesting = False)

    return graphs
```

```python
from eden.util import display
import itertools as it
graphs = pre_process(rfam_url(rfam_id))
for graph in it.islice(graphs,start,stop,step):
    display.draw_graph(graph, vertex_color='weight', colormap='autumn', **opts)
```

Vectorization setup

```python
from eden.graph import Vectorizer
vectorizer = Vectorizer( r=3, d=5 )
n_jobs = 1
```

once the sequences are transformed into vectors, clustering and low dimnsional embedding become possible through standard ML libraries

```python
%%time
from eden.modifier.fasta import fasta_to_fasta, one_line_modifier
headers = fasta_to_fasta( rfam_url( rfam_id ), modifier=one_line_modifier, header_only=True )

import itertools
num = 50
start = 0
stop = start + num

labels = [label for label in itertools.islice( headers, start, stop )]
graphs = itertools.islice( pre_process( rfam_url( rfam_id ) ), start, stop )

from eden.util.display import embed2D
embed2D( graphs, vectorizer, labels=labels, size=size, gamma=80 , colormap='YlGnBu' )
```

prepare a function that composes all design actions over the original data

```python
def pre_process( data ):
    from eden.converter.rna.rnashapes import rnashapes_to_eden
    graphs = rnashapes_to_eden( data, shape_type = 5, energy_range = 35, max_num = 3 )
    return graphs
```

employ the vectorizer to transform the graph instances into sparse vectors and collect all the vectors in a data matrix 

```python
%%time
graphs = pre_process( rfam_url( rfam_id ) )
X1 = vectorizer.transform( graphs, n_jobs = n_jobs )
print 'Instances: %d ; Features: %d with an avg of %d features per instance' % (X1.shape[0], X1.shape[1],  X1.getnnz()/X1.shape[0])
```

repeat the procedure on a different RNA family that will be used as negative class

...or randomly shuffle the sequences to create the negative examples

```python
#make custom modifier: assume as attributes 'header' and 'seq' and other options
def shuffle_modifier(header = None, seq = None, **options):
    times =  options.get('times',1)
    order =  options.get('order',1)
    for i in range(times):
        kmers = [ seq[i:i+order] for i in range(0,len(seq),order) ]
        random.shuffle(kmers)
        seq_out = ''.join(kmers)
        yield header
        yield seq_out
```

```python
%%time
from eden.modifier.fasta import fasta_to_fasta
graphs = pre_process( fasta_to_fasta( rfam_url( rfam_id ) , modifier=shuffle_modifier, times=2, order=2) )
X2 = vectorizer.transform( graphs, n_jobs = n_jobs )
print 'Instances: %d ; Features: %d with an avg of %d features per instance' % (X2.shape[0], X2.shape[1],  X2.getnnz()/X2.shape[0])
```

we concatenate the two data matrices into one matrix
and we create a target vector that identifies with +1 and -1 the instances comping form the orginal and the shuffled data

Fit a predictor

```python
%%time
from eden.util import estimate_predictive_performance
estimator = estimate_predictive_performance( positive_data_matrix=X1, negative_data_matrix=X2, cv=3 )
```

```python
from eden.graph import Annotator
annotator = Annotator( estimator = estimator, vectorizer = vectorizer )
```
