```python
from eden import graph
from eden.util import io
from eden.converters import gspan
```

```python
max_radius=6
max_distance=6
input_data_url='http://www.bioinf.uni-freiburg.de/~costa/bursi300.gspan'
input_target_url='http://www.bioinf.uni-freiburg.de/~costa/bursi300.target'
```

```python
%%time
vec=graph.Vectorizer(r=max_radius,d=max_distance)
g_it=gspan.gspan_to_eden(input_data_url)
X=vec.transform(g_it)

print 'Instances: %d Features: %d with an avg of %d features per instance' % (X.shape[0], X.shape[1],  X.getnnz()/X.shape[0])
```

```python
y=io.load_target(input_target_url)
print('Target size:%d' % y.shape[0])
print('Target classes:%d' % len(set(y)))
```

```python
%%time
from sklearn.neighbors import kneighbors_graph

num_neighbours=2

A = kneighbors_graph(X, num_neighbours)
```

```python
%%time

from eden.util import display

display.draw_adjacency_graph(A,y,node_size=80)
```

```python
from sklearn import metrics

K=metrics.pairwise.pairwise_kernels(X, metric='linear')
print K
```
