```python
from eden.converter.RNA import RNA_PLFOLD

#RFAM id of a family
rf='RF00005'
#we compose the correspondent URL to fetch data from RFAM database
url_string='http://rfam.xfam.org/family/%s/alignment?acc=%s&format=fastau&download=0'%(rf,rf)

RNAplfold_options = {'window_size' : 150, 'max_bp_span' : 100, 'avg_bp_prob_cutoff' : 0.1, 'max_num_edges' : 3}

graphs = RNA_PLFOLD.RNA_PLFOLD_to_eden(input = url_string, input_type = 'url' , **RNAplfold_options )


# Ouput some graphs.
num = 5

from eden.util import display
import itertools as it
for g in it.islice(graphs,num):
    display.draw_graph(g, size=20, node_border = False, node_size = 400, 
                       secondary_vertex_label = 'position', secondary_edge_label = 'value')
```
